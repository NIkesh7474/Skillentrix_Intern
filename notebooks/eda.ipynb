# Exploratory Data Analysis - Telco Customer Churn

This notebook performs EDA focusing on churn distribution and key relationships: Contract, Payment Method, Tenure, Monthly Charges, and correlations. Place dataset at `data/telco_customer_churn.csv`.

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")
%matplotlib inline
```

```python
# Load data (adjust path if needed)
df = pd.read_csv('../data/telco_customer_churn.csv')
df.head()
```

```python
# Basic info and missing values
df.info()
print('Missing values:
', df.isna().sum())
```

```python
# Ensure TotalCharges numeric
if 'TotalCharges' in df.columns:
    df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Encode churn for easy plotting
if 'Churn' in df.columns:
    df['Churn_flag'] = df['Churn'].map({'Yes': 1, 'No': 0})

df[['Churn', 'Churn_flag']].value_counts()
```

## Churn Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Churn')
plt.title('Churn Distribution')
plt.show()
```

## Contract Type vs Churn

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Contract', hue='Churn')
plt.title('Contract Type vs Churn')
plt.show()
```

## Payment Method vs Churn

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='PaymentMethod', hue='Churn')
plt.xticks(rotation=45)
plt.title('Payment Method vs Churn')
plt.show()
```

## Tenure vs Churn

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data=df, x='tenure', hue='Churn', common_norm=False, fill=True)
plt.title('Tenure distribution by Churn')
plt.show()
```

## Monthly Charges vs Churn

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Churn', y='MonthlyCharges')
plt.title('Monthly Charges by Churn')
plt.show()
```

## Correlation Heatmap (numeric features)

```python
num_cols = df.select_dtypes(include=["number"]).columns.tolist()
plt.figure(figsize=(10,8))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```
